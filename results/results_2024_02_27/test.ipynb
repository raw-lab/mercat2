```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

jf = pd.read_csv('jf-sorted.tsv', sep='\t')
mc = pd.read_csv('mercat2-sorted.tsv', sep='\t')
km = pd.read_csv('kmc-sorted.tsv', sep='\t')

print(jf)
print(mc)
print(km)

df = pd.DataFrame({'Jellyfish': jf.iloc[:, 1], 'KMC': km.iloc[:, 1], 'Mercat2': mc.iloc[:, 1]})
```

```python
corr = df.corr(method = 'spearman')

sns.heatmap(corr, annot = True)

plt.savefig("plot.png", dpi=300)
plt.savefig("plot.jpg", dpi=300)
plt.show()
```
